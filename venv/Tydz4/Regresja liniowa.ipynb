```python
#wczytanie danych
import pandas
df = pandas.read_csv("diamonds_single.csv", sep=";", header=0)
df.head()
df.shape
```

```python
#wizualizacja danych
%matplotlib inline
df.plot.scatter(x="carat", y="price")
```

```python
#podział danych na zbiór danych treningowych i testowych
from sklearn.model_selection import train_test_split
X_diam = df[["carat"]]
Y_diam = df[["price"]]
XTrain, XTest, YTrain, YTest = train_test_split(X_diam, Y_diam, random_state=1)
```

```python
#przygotowanie modelu
from sklearn import linear_model
regr = linear_model.LinearRegression()
display(regr)
```

```python
#trenowanie 
regr.fit(XTrain, YTrain)
```

```python
#predykcja na danych testowych
y_pred = regr.predict(XTest)
```

```python
#wyniki
#wzor na funkcje liniowa y= ax + b
from sklearn.metrics import mean_squared_error
print("Wspolczynnik a: ", regr.coef_)
print("Wyraz wolny b:", regr.intercept_)
print("Blad srednikwadratowy", mean_squared_error(YTest, y_pred))
print("Wspolczynnik determinacji r2:", regr.score(XTrain, YTrain))
```

```python
import matplotlib.pyplot as plt
plt.scatter(XTrain, YTrain)
plt.plot(XTest, y_pred, color="red", linewidth=3)
```
