```python
#import danych
import pandas
df = pandas.read_csv("diamonds_multi.csv",header = 0, sep = ";")
df.head()
df['color'].unique()

df['cut'].unique()
```

```python
#przygotowanie danych
colorMap = {
    'D':7,
    'E':6,
    'F':5,
    'G':4,
    'H':3,
    'I':2,
    'J':1
}
cutMap = {
    'Ideal':5,
    'Premium':4,
    'Very Good':3,
    'Good':2,
    'Fair':1
}
df['colorNumber'] = df['color'].map(colorMap)
df['cutNumber'] = df['cut'].map(cutMap)
newDf = df.drop(['color', 'cut'], axis = 1)
display(newDf)
```

```python
#podział danych na testowe i do nauki
from sklearn.model_selection import train_test_split
x_diam = newDf[['carat','x','y','z','cutNumber','colorNumber']]
y_diam = newDf[['price']]
Xtrain, Xtest, Ytrain, Ytest = train_test_split(x_diam, y_diam, random_state=1)
```

```python
#przygotowanie modelu
from sklearn import linear_model
regr = linear_model.LinearRegression()
```

```python
#trenowanie modelu 
regr.fit(Xtrain, Ytrain)
```

```python
#predykcja 
y_pred = regr.predict(Xtest)
```

```python
#wyniki 

from sklearn.metrics import mean_squared_error
# y = a1x1 + a2x2 + a3x3 ... + b 
print('Współczynniki: ', regr.coef_)
print('Wyraz wolny: ', regr.intercept_)
print('Błąd średniokwadratowy: %.2f' % mean_squared_error(Ytest, y_pred))
print('Współczynnik determinacji R^2: %.2f' % regr.score(Xtrain, Ytrain))
```
